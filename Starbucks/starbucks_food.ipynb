## Convert column to numeric

```python
starbucks.columns = starbucks.columns.str.strip()
starbucks
```

```python
print(starbucks.columns.tolist())
```

```python
cols = ['Calories', 'Total Fat (g)', 'Trans Fat (g)', 'Saturated Fat (g)',
                  'Sodium (mg)', 'Total Carbohydrates (g)', 'Cholesterol (mg)',
                  'Dietary Fibre (g)', 'Sugars (g)', 'Protein (g)', 
                  'Caffeine (mg)']

for col in cols:
    starbucks[col] = pd.to_numeric(starbucks[col], errors='coerce')
```

```python
print(starbucks[cols].isnull().sum())
```

```python
percent_cols = ['Vitamin A (% DV)', 'Vitamin C (% DV)', 'Calcium (% DV)', 'Iron (% DV)']

for col_name in percent_cols:
    starbucks[col_name] = starbucks[col_name].astype(str).str.replace('%', '', regex=False)
    starbucks[col_name] = pd.to_numeric(starbucks[col_name], errors='coerce')
```

## Fill with column mean

```python
starbucks['Caffeine (mg)'] = starbucks['Caffeine (mg)'].fillna(starbucks['Caffeine (mg)'].mean())
```

```python
starbucks['Total Fat (g)'] = starbucks['Total Fat (g)'].fillna(starbucks['Total Fat (g)'].mean())
```

```python
starbucks.isnull().sum()
```

## What is the average nutritional profile across all drinks? 

```python
col = [
    'Calories',
    'Sugars (g)',
    'Total Fat (g)',
    'Trans Fat (g)',
    'Saturated Fat (g)',
    'Sodium (mg)',
    'Total Carbohydrates (g)',
    'Cholesterol (mg)',
    'Dietary Fibre (g)',
    'Protein (g)',
    'Vitamin A (% DV)',
    'Vitamin C (% DV)',
    'Calcium (% DV)',
    'Iron (% DV)',
    'Caffeine (mg)'

]
avg = starbucks[col].mean()
print(avg)
```

## Which drinks have the highest and lowest calories?

```python
# Highest calorie drinks (top 5)
highest_cal = starbucks[['Beverage', 'Beverage_prep', 'Calories']].sort_values(by='Calories', ascending=False).head(5)

# Lowest calorie drinks (bottom 5, nonzero)
lowest_cal = starbucks[starbucks['Calories'] > 0][['Beverage', 'Beverage_prep', 'Calories']].sort_values(by='Calories').head(5)

print("Highest Calorie Drinks:")
print(highest_cal.to_string(index=False))

print("\nLowest Non-Zero Calorie Drinks:")
print(lowest_cal.to_string(index=False))
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=highest_cal, x='Calories', y='Beverage', hue='Beverage_prep', palette='Reds')
plt.title(' Highest Calorie Drinks')
plt.xlabel('Calories')
plt.ylabel('Beverage')
plt.show()
```

## Do calories correlate with sugar? (Calories vs. Sugars)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=starbucks, x='Sugars (g)', y='Calories', hue='Beverage_category', alpha=0.7)
plt.title('Calories vs. Sugar Content in Starbucks Drinks')
plt.xlabel('Sugar (g)')
plt.ylabel('Calories')
plt.legend(title='Beverage Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
cor = starbucks['Calories'].corr(starbucks['Sugars (g)'])
print(f"Correlation between Calories and Sugar: {cor:.2f}")
```

We can see that there is a positive correlation between calories and sugar content in Starbucks drinks, indicating that as the sugar content increases, the calorie count tends to increase as well. This is a common trend in many food and beverage products, where higher sugar levels contribute to higher caloric values. The correlation between Calories and Sugar is approximately 91%, indicating a very strong positive relationship.

This implies that sugar is one of the main contributors to calories in these drinks. For example, sugary drinks like White Chocolate Mocha or Java Chip Frappuccino rank high on both sugar and calorie scales. In contrast, drinks such as Brewed Coffee and Caffe Americano, which contain little or no sugar, are also low in calories.

## Are drinks with more fat also higher in calories?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=starbucks, x='Total Fat (g)', y='Calories', hue='Beverage_category', alpha=0.7)
plt.title('Calories vs. Total Fat in Starbucks Drinks')
plt.xlabel('Total Fat (g)')
plt.ylabel('Calories')
plt.legend(title='Beverage Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
fat_cal_corr = starbucks['Calories'].corr(starbucks['Total Fat (g)'])
print(f"Correlation between Calories and Total Fat: {fat_cal_corr:.2f}")
```

The correlation between Calories and Total Fat (g) is approximately 63%, indicating a moderate positive relationship. This suggests that drinks with higher fat content tend to have more calories

## Which drinks are highest in Vitamin A, Vitamin C, Calcium, Iron?

```python
vitamin_a_top = starbucks.loc[starbucks['Vitamin A (% DV)'].idxmax()]
vitamin_c_top = starbucks.loc[starbucks['Vitamin C (% DV)'].idxmax()]
calcium_top   = starbucks.loc[starbucks['Calcium (% DV)'].idxmax()]
iron_top      = starbucks.loc[starbucks['Iron (% DV)'].idxmax()]

summary = pd.DataFrame({
    'Nutrient': ['Vitamin A', 'Vitamin C', 'Calcium', 'Iron'],
    'Beverage': [
        vitamin_a_top['Beverage'],
        vitamin_c_top['Beverage'],
        calcium_top['Beverage'],
        iron_top['Beverage']
    ],
    'Preparation': [
        vitamin_a_top['Beverage_prep'],
        vitamin_c_top['Beverage_prep'],
        calcium_top['Beverage_prep'],
        iron_top['Beverage_prep']
    ],
    'Highest % DV': [
        vitamin_a_top['Vitamin A (% DV)'],
        vitamin_c_top['Vitamin C (% DV)'],
        calcium_top['Calcium (% DV)'],
        iron_top['Iron (% DV)']
    ]
})

print(summary)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(summary['Nutrient'], summary['Highest % DV'], color=['orange', 'green', 'blue', 'brown'])

# Add labels
for bar, label in zip(bars, summary['Beverage']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 2, label, ha='center', fontsize=9)

plt.title('Drinks with the Highest % Daily Value for Each Nutrient')
plt.ylabel('% Daily Value')
plt.show()
```
